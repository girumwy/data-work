This is an example of data inspection and analysis I was tasked to deal with while working as an admin intern for Alama Onboarding which is a small business in Ethiopia that helps students with college prep. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data:
df = pd.read_csv('students.csv')
```

```python
# example of first 5 rows of dataset for reference:
df.head()
```

```python
# inspection, cleaning, and analysis of the data:
print(df.head())
print(df.isnull().sum())

df.columns = df.columns.str.replace(' ', '_')

avg_math = df['math_score'].mean()
avg_reading = df['reading_score'].mean()
avg_writing = df['writing_score'].mean()

print('Average math score:', avg_math)
print('Average reading score:', avg_reading)
print('Average writing score:', avg_writing)

# basic statistics by group:
grouped = df.groupby("gender")[["math_score", "reading_score", "writing_score"]].mean()
print(grouped)

# descriptive statistics:
print(df[["math_score", "reading_score", "writing_score"]].describe())


# saving cleaned file for future use:
df.to_csv('students_cleaned.csv', index=False)
```

```python
# visualizing the data:
df["math_score"].hist(bins=10)
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
